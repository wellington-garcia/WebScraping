```python
# importando bibliotecas
import time
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

# Categoria - Livros

```python
#declarando variáveis cards
cards = []

for i in range(12):
    time.sleep(10)
    url = 'https://www.magazineluiza.com.br/busca/livros%20de%20fic%C3%A7%C3%A3o/?page=' + str(i + 1) + '&sortOrientation=desc&sortType=soldQuantity'
    headers = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/94.0.4606.81 Safari/537.36'}
    response = requests.get(url, headers = headers)
    soup = BeautifulSoup(response.content, "html.parser")

    anuncios = soup.find('ul', class_="sc-iktFfs ekIdSn sc-irlQje gbimYp sc-fBxREx fjFlfr").findAll('li', class_="sc-eCVOVf loRbcV")
    
    #Obtendo as informações dos cards
    for anuncio in anuncios:
        card = {}

        #nome do livro
        nome = anuncio.find('h2', class_="sc-bHXGc hsFKpx")
        try:
            card['Livro'] = nome.getText().split('-', 1)[1]
        except IndexError:
            card['Livro'] = nome.getText()

        #preco do livro
        try:
            card['preco'] = anuncio.find('p').getText().replace('\xa0', ' ').split()[1]
        except AttributeError:
            card['preco'] = 0

        #preco com desconto
        try:
            card['preco_desconto'] = anuncio.find('p', class_="sc-hKgJUU kegCEa sc-bxnjHY cTpdOW").getText().replace('\xa0', ' ').split()[1]
        except AttributeError:
            card['preco_desconto'] = 0
        
#         card['preco_desconto'] = anuncio.find('p', class_="sc-hKgJUU kegCEa sc-bxnjHY cTpdOW").getText()

        #capiturando o preco parcelaado
        try:
            value = anuncio.find('div', class_="sc-jSgvzq fdbhJo sc-kGNyvp kBUBND").findAll('p')[2].getText().split()[1:5:3]   
            card['parcelas'] = value[0].split('x')[0]
            card['valor_parcela'] = value[1]
        except IndexError:
            card['parcelas'] = 1
            card['valor_parcela'] = card['preco_desconto']
        except AttributeError:
            card['parcelas'] = "0"
            card['valor_parcela'] = 0

        #Adicionando resultado a lista cards
        cards.append(card)
    
#Criando um dataframe com os resultados
dataset = pd.DataFrame(cards)
dataset.to_csv('./livros_magalu.csv', sep=';', index = False, encoding = 'utf-8-sig')
dataset
```

### Função para tratar ps preços com virgulas

```python
def tratamento_virgula_ponto(dataset, colun):
    for i in range(dataset.shape[0]):
        dataset.loc[i, colun] = dataset.loc[i, colun].replace('.', '')
        dataset.loc[i, colun] = dataset.loc[i, colun].replace(',', '.')  
    return dataset.astype({ colun : float})
```
